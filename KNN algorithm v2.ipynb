```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

%pylab inline --no-import-all

import pandas as pd
```

```python
df = pd.read_csv("ratings.csv", dtype='str')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(df, test_size=1/3, random_state=999)

print(X_train.shape, X_val.shape)
print(y_train.shape, y_val.shape)
```

```python
knn31 = KNeighborsClassifier(n_neighbors=3, p=1)
knn31.fit(X_train, y_train)

y_pred_knn31 = knn31.predict(X_val)
print(accuracy_score(y_val, y_pred_knn31))

print(classification_report(y_val, y_pred_knn31))
print(confusion_matrix(y_val, y_pred_knn31))
```

```python
knn32 = KNeighborsClassifier(n_neighbors=3, p=2)
knn32.fit(X_train, y_train)

y_pred_knn32 = knn32.predict(X_val)
print(accuracy_score(y_val, y_pred_knn32))

print(classification_report(y_val, y_pred_knn32))
print(confusion_matrix(y_val, y_pred_knn32))
```
